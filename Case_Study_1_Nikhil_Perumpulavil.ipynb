# Step 3: Plot a bar graph of male vs female buyers participated in the sales

```python
Y = data['Buyer Gender'].value_counts()
Y
```

```python
plt.figure(figsize = (8,8))
sns.countplot(data['Buyer Gender'])
plt.title('Bar graph of male vs female buyers', fontsize = 14)
plt.xticks()
plt.yticks()
plt.grid()
```

5052 Females and 4948 males participated in the sales

# Step 4: Find the top 5 cars based on their sales price

```python
sorted_SP_df = SP_df.sort_values('Sale Price', ascending = False)
sorted_SP_df
```

```python
sorted_SP_df.head()
```
