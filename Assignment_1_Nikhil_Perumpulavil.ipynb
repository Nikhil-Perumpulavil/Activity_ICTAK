# Step 5. Perform meaningful visualizations using the dataset. Bring at least 3 visualizations

# Pair plot

```python
sns.pairplot(data)
```

# Scatter Plot

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
plt.scatter(data['SL'], data['SW'],s=12, marker='^', c='red')
plt.title('Scatter plot of SL vs SW', fontsize=14)
plt.xlabel('SL', fontsize=12)
plt.ylabel('SW', fontsize=12)

plt.subplot(2,1,2)
plt.scatter(data['PL'], data['PW'],s=12, marker='*', c='green')
plt.title('Scatter plot of PL vs PW', fontsize=14)
plt.xlabel('PL', fontsize=12)
plt.ylabel('PW', fontsize=12)

plt.tight_layout()
```

# Box plot

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
sns.boxplot(x = 'Classification', y = 'PL', data=data)
plt.title('Box plot of Classification vs PL', fontsize=14)
plt.xlabel('Classification', fontsize=12)
plt.ylabel('PL', fontsize=12)

plt.subplot(2,1,2)
sns.boxplot(x = 'Classification', y = 'SL', data=data)
plt.title('Box plot of Classification vs SL', fontsize=14)
plt.xlabel('Classification', fontsize=12)
plt.ylabel('SL', fontsize=12)

plt.tight_layout()
```

# Violin plot

```python
plt.figure(figsize=(8,6))

plt.subplot(2,1,1)
sns.violinplot(x = 'Classification', y = 'PL', data=data)
plt.title('Violin plot of Classification vs PL', fontsize=14)
plt.xlabel('Classification', fontsize=12)
plt.ylabel('PL', fontsize=12)

plt.subplot(2,1,2)
sns.violinplot(x = 'Classification', y = 'SL', data=data)
plt.title('Violin plot of Classification vs SL', fontsize=14)
plt.xlabel('Classification', fontsize=12)
plt.ylabel('SL', fontsize=12)

plt.tight_layout()
```
